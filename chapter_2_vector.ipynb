# Vector 
Create a 2D vector v, 10 scalars that are drawn at random from a normal (Gaussian) distribution, and plot all 10 scalar-vector multiplications on top of each other. What do you notice?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scalars = np.random.uniform(size=(10,1))
scalars
```

```python
vec = np.array([1, 1])
vec
```

```python
plt.plot([0,vec[0]],[0,vec[1]])
```

```python
plt.plot([0,vec[0]],[0,vec[1]])
for scalar in scalars:
    plt.plot([0,vec[0]*scalar],[0,vec[1]*scalar])
```

### Answer:
all vectors are scaled into different magnitude by have a same direction.
